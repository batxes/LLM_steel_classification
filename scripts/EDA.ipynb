```python
import pandas as pd
from os import listdir
from os.path import isfile, join
import re

path_to_data = "../resources/"

onlyfiles = [f for f in listdir(path_to_data) if isfile(join(path_to_data, f))]
datafiles = [f for f in onlyfiles if f.lower().endswith(".xlsx")]

print (f"Data files are: {datafiles}")
```

```python
dfs = []
file_dict = dict()
for datafile in datafiles:
    xlsx = pd.ExcelFile(path_to_data + datafile)
    print (f"Sheet names: {xlsx.sheet_names}")
    for n,sheet_name in enumerate(xlsx.sheet_names):
        name = datafile + f"_{n}"
        good_header = False
        header = 0
        while not good_header:
            df = pd.read_excel(xlsx, sheet_name, header=header)
            if any(col.startswith('Unnamed') for col in df.columns):
                header += 1
            else:
                print (f"Column names: {list(df.columns)}")
                good_header = True
                dfs.append(df)
                file_dict[name] = df
    print ()
    
for key, value in file_dict.items():
    print (f" {key} contains {len(value)} sheet inside.")
    print (value)
```

Some files contain multiple tables. Get indices of NaN rows and extract each table as df

```python
for key, df in file_dict.items():
    print ("File: {}".format(key))
    nan_rows = df[df.isna().all(axis=1)]
    if len(nan_rows != 0):
        aux_df_list = []
        print ("--> Contains multiple dfs inside")
        nan_indexes = nan_rows.index
        start_index = 0
        for index in nan_indexes:
            aux_df = df.iloc[start_index:index].reset_index(drop=True)
            #print (len(aux_df))
            #if not (aux_df.iloc[0].isna().all()): #check if first row of values if full of NaNs. Get some problems with headers...
            if len(aux_df) > 1:
                aux_df_list.append(aux_df)
            start_index = index
        # add last table:
        aux_df = df.iloc[start_index:]
        aux_df_list.append(aux_df)
        file_dict[key] = aux_df_list
    
```

```python
def extract_dimensions(material):
    dimensions = re.findall(r"(\d+[,|\.*\d*]*)\s*[m]*\s*x\s*(\d+[,|\.]*\d*)\s*[m]*\s*", material)
    return dimensions[0] if dimensions else (None, None)
    
def remove_dimensions(material):
    description = re.sub(r"(\d+[,|\.*\d*]*)\s*[m]*\s*x\s*(\d+[,|\.]*\d*)\s*[m]*\s*","", material)
    return description
```

```python
df = new_df_source1_sheet_0.copy()
df[['Thickness', 'Width']] = df['Material'].apply(lambda x: pd.Series(extract_dimensions(x)))
df['Material'] = df['Material'].apply(lambda x: pd.Series(remove_dimensions(x)))
```
